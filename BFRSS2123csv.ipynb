```python
import pandas as pd
```

```python
df = pd.read_csv("BRFSS_2021_2024_diabetes_ready_v2.csv")  # or pd.read_excel(), pd.read_sas(), etc.

df['year'] = pd.to_numeric(df['year'], errors='coerce')

df_filtered = df[df['year'].isin([2021, 2023])]

print(df_filtered.head())
print(f"dataset shape: {df_filtered.shape}")
```

```python
df_filtered = df_filtered.drop(columns=['sleep_hours', 'cholesterol_check'], errors='ignore')


print(df_filtered.head())
print(f" Columns after drop: {df_filtered.columns.tolist()}")
```

```python
for col in df.columns:
    print(f"\nColumn: {col}")
    percentages = df[col].value_counts(normalize=True) * 100  # convert to %
    print(percentages.round(2))  # round to 2 decimals
```

```python
df_filtered.to_csv("only2123.csv", index=False)
print("Filtered data saved to filtered_data.csv")
```
